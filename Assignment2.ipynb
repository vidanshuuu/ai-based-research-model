```python
from langchain_community.tools import DuckDuckGoSearchRun, WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
from transformers import pipeline
```

```python
search = DuckDuckGoSearchRun()
wiki = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())
```

```python
generator = pipeline(
    "text-generation",
    model="distilgpt2",
    max_new_tokens=200
)
```

```python
def research_topic(topic):
    web_data = search.run(topic)
    wiki_data = wiki.run(topic)

    # keep data short to avoid crash
    return f"""
    Web Summary:
    {web_data[:500]}

    Wikipedia Summary:
    {wiki_data[:500]}
    """
```

```python
def generate_report(topic, data):
    prompt = f"""
Create a structured report on: {topic}

Include:
- Cover Page
- Title
- Introduction
- Key Findings
- Challenges
- Future Scope
- Conclusion

Data:
{data}
"""

    result = generator(prompt)[0]["generated_text"]
    return result
```

```python
topics = [
    "Impact of AI in Healthcare",
    "Role of AI in Education"
]

for topic in topics:
    data = research_topic(topic)
    report = generate_report(topic, data)

    print("\n\n==============================\n")
    print(report)
```

```python
topics = [
    "Impact of AI in Healthcare",
    "Role of AI in Education"
]

for topic in topics:
    print("\n\n==============================")
    print(f"   RESEARCH REPORT FOR: {topic}")
    print("==============================\n")

    try:
        data = research_topic(topic)
        report = generate_report(topic, data)
        print(report)
    except Exception as e:
        print("Error occurred:", e)
```
